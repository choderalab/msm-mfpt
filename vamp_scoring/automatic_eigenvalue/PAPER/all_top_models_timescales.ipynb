```python
import pyemma
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('ticks')
sns.set_context('paper', font_scale=1.5)
import numpy as np
from glob import glob
```

```python
pwd
```

```python
dtrajs_chignolin = list(np.load('../chignolin/chignolin_dtrajs_lag10ns.npy'))
```

```python
dtrajs_villin = list(np.load('../villin/dtrajs_villin_lag10ns.npy'))
```

```python
dtrajs_trpcage = list(np.load('../trpcage/dtrajs_trpcage_lag10ns.npy'))
```

```python
dtrajs_ntl9 = list(np.load('../ntl9/dtrajs_ntl9_lag10ns.npy'))
```

```python
its_chignolin = pyemma.msm.its(dtrajs_chignolin, 2000, errors='bayes')
```

```python
its_villin = pyemma.msm.its(dtrajs_villin, 2000, errors='bayes')
```

```python
its_trpcage = pyemma.msm.its(dtrajs_trpcage, 2000, errors='bayes')
```

```python
its_ntl9 = pyemma.msm.its(dtrajs_ntl9, 2000, errors='bayes')
```

```python
f, ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7, ax8)) = plt.subplots(4, 2, figsize=(10,10))

pyemma.plots.plot_implied_timescales(its_chignolin, ax=ax1, ylog=False, dt=0.2, units='ns')
pyemma.plots.plot_implied_timescales(its_chignolin, ax=ax2, ylog=True, dt=0.2, units='ns')

pyemma.plots.plot_implied_timescales(its_villin, ax=ax3, ylog=False, dt=0.2, units='ns')
pyemma.plots.plot_implied_timescales(its_villin, ax=ax4, ylog=True, dt=0.2, units='ns')

pyemma.plots.plot_implied_timescales(its_trpcage, ax=ax5, ylog=False, dt=0.2, units='ns')
pyemma.plots.plot_implied_timescales(its_trpcage, ax=ax6, ylog=True, dt=0.2, units='ns')

pyemma.plots.plot_implied_timescales(its_ntl9, ax=ax7, ylog=False, dt=0.2, units='ns')
pyemma.plots.plot_implied_timescales(its_ntl9, ax=ax8, ylog=True, dt=0.2, units='ns')

ax1.text(-0.6, 0.5, 'Chignolin', transform=ax1.transAxes, fontweight='bold')
ax3.text(-0.6, 0.5, 'Villin', transform=ax3.transAxes, fontweight='bold')
ax5.text(-0.6, 0.5, 'Trp-cage', transform=ax5.transAxes, fontweight='bold')
ax7.text(-0.6, 0.5, 'NTL9', transform=ax7.transAxes, fontweight='bold')

ax1.get_yaxis().set_label_coords(-0.175,0.5)
ax3.get_yaxis().set_label_coords(-0.175,0.5)
ax5.get_yaxis().set_label_coords(-0.175,0.5)
ax7.get_yaxis().set_label_coords(-0.175,0.5)

ax1.set_ylabel('timescale / ns', fontweight='bold')
ax3.set_ylabel('timescale / ns', fontweight='bold')
ax5.set_ylabel('timescale / ns', fontweight='bold')
ax7.set_ylabel('timescale / ns', fontweight='bold')

ax2.set_ylabel('')
ax4.set_ylabel('')
ax6.set_ylabel('')
ax8.set_ylabel('')

ax7.set_xlabel('lag time / ns', fontweight='bold')
ax8.set_xlabel('lag time / ns', fontweight='bold')

ax1.set_xlabel('')
ax2.set_xlabel('')
ax3.set_xlabel('')
ax4.set_xlabel('')
ax5.set_xlabel('')
ax6.set_xlabel('')

ax1.axvline(150, c='green')
ax2.axvline(150, c='green')

ax3.axvline(100, c='green')
ax4.axvline(100, c='green')

ax5.axvline(100, c='green')
ax6.axvline(100, c='green')

ax7.axvline(200, c='green')
ax8.axvline(200, c='green')

ax1.axhline(2039.0/5, c='green')
ax2.axhline(2039.0/5, c='green')

ax3.axhline(5802.3/5, c='green')
ax4.axhline(5802.3/5, c='green')

ax5.axhline(14245.7/5, c='green')
ax6.axhline(14245.7/5, c='green')

ax7.axhline(96761.6/5, c='green')
ax8.axhline(96761.6/5, c='green')

plt.tight_layout()
plt.savefig('all_models_its_400ns.pdf', bbox_inches='tight')
plt.savefig('all_models_its_400ns.png', bbox_inches='tight', dpi=300)
```
